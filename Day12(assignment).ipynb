```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'banking_data.csv'
banking_data = pd.read_csv(file_path)
```

```python
# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_per_account_type = banking_data.groupby('Account_Type')['Transaction_Amount'].sum()
total_transaction_per_account_type.plot(kind='bar', color='skyblue', title='Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.show()
```

```python
# Task 2: Create a pie chart to show the percentage of transactions per Branch
transaction_per_branch = banking_data['Branch'].value_counts()
transaction_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Percentage of Transactions per Branch')
plt.ylabel('')
plt.show()
```
