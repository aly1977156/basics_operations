```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'pharma_data.csv'  # Update this path with your actual CSV file
pharma_data = pd.read_csv(file_path)
```

```python
# Data cleaning: Check for missing values and duplicates
print(pharma_data.isnull().sum())  # Checking for missing values
pharma_data = pharma_data.drop_duplicates()  # Removing duplicates
```

```python
# Visualizations

# 1. Bar plot showing total sales per region
total_sales_per_region = pharma_data.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 6))
total_sales_per_region.plot(kind='bar', color='skyblue', title='Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Scatter plot to visualize the relationship between Marketing_Spend and Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=pharma_data, color='purple')
plt.title('Relationship between Marketing Spend and Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()
```

```python
# 3. Boxplot comparing drug effectiveness across different age groups
plt.figure(figsize=(6, 2))
sns.boxplot(x='Age_Group', y='Effectiveness', data=pharma_data, palette='Set2')
plt.title('Drug Effectiveness across Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.show()
```
