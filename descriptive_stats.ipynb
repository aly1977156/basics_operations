#### Visualization using boxplot

```python
import statistics as sts
import pandas as pd


a = [15, 20, 10, 30, 35, 25,60,170,150]
sc=pd.Series(a)
print(sc)


plt.boxplot(sc)


med = sts.median(a)

plt.title(f"Scores\n Median={med}")

```
